```python
adstock = np.empty_like(b1_normalised_grps)
```

```python
adstock[0] = b1_normalised_grps[0]
```

```python


import statsmodels.api as sm

# Define the independent and dependent variables
X = df[[ 'TP1','TP3', 'CP2']]
y = df['B1_Sales_Percent_of_Category']

# Normalize the independent variables
X = (X - X.mean()) / X.std()

# Add a constant to the independent variables (intercept term)
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python
import matplotlib.pyplot as plt

# Assuming 'model' is the fitted statsmodels OLS model from the previous code
coef_values = model.params.drop('const')  # Exclude the constant term


plt.figure(figsize=(10, 6))
plt.bar(coef_values.index, coef_values.values)
plt.xlabel("Coefficients")
plt.ylabel("Coefficient Values")
plt.title("Bar plot of Coefficients")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Define features (independent variables) and target (dependent variable)
features = ['adstockB1', 'NonTVAdstockB1', 'TP1', 'TP3', 'CP1', 'CP2']
pca_features = ['B1 Number of Dealers', 'B1 Avg Price per Litre', 'B1 WTDL Distn']
target = 'B1_Sales_Percent_of_Category'

# Create a subset of the DataFrame with selected features and target
subset = df[features + pca_features + [target]]

# Handle missing values (you might need to adjust the strategy)
subset.fillna(subset.mean(), inplace=True) # Replace missing with the mean

# Normalize the features
scaler = MinMaxScaler()
subset[features + pca_features] = scaler.fit_transform(subset[features + pca_features])

# Apply PCA
pca = PCA(n_components=1)  # Reduce to one principal component
subset['PCA_Component'] = pca.fit_transform(subset[pca_features])

# Prepare the data for regression
X = subset[features + ['PCA_Component']]
y = subset[target]

# Add a constant to the independent variables (intercept term)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```

```python


import matplotlib.pyplot as plt

# Extract p-values from the model summary
p_values = model.pvalues[1:]  # Exclude the constant term

# Create a bar plot of p-values
plt.figure(figsize=(10, 6))
plt.bar(p_values.index, p_values.values)
plt.xlabel("Predictors")
plt.ylabel("P-values")
plt.title("P-values for Regression Coefficients")
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
